```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

## create a line plot
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic line plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

## create a customize line plot

plt.plot(x,y,color= 'red', linestyle= '--', marker= 'o', linewidth=4, markersize = 9)
plt.grid(True)
```

```python
## Multiple plots
## Sample data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
z = [1, 2, 3, 4, 5]

plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(x, y, color = 'green')
plt.title('Plot 1')

plt.subplot(2,2,2)
plt.plot(y, x, color = 'red')
plt.title('Plot 2')

plt.subplot(2,2,3)
plt.plot(x, z, color = 'blue')
plt.title('Plot 3')


plt.subplot(2,2,4)
plt.plot(z, y, color = 'yellow')
plt.title('Plot 4')
```

```python
## Bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 18, 6]

# create a bar plot
plt.bar(categories, values)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Grade Evalution of Student")
plt.show()
```

## Histograms

Histograms are used to represent the distribution of dataset.They divide the data into bins and count the number of data points in each bin.

```python
## Samples data
data = [1, 2, 3, 4, 5, 6, 4, 4, 4, 5, 5, 5, 3, 3, 3, 6, 6, 6, 2, 4, 4, 4]

## create histograms
plt.hist(data, bins=5, color= 'orange',edgecolor= 'black' )
```
